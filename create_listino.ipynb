```python
import pandas as pd
import numpy as np
```

```python
df_veneziane = pd.read_excel('data_raw/Aggiornamento_ART_COSTI_per_tipo_articoli_21-08_2023.xlsx', sheet_name='art_costi_veneziane')
df_veneziane['Codice_variante'] = df_veneziane['COD_ART_VAR_TIPO_COS_AZ_STDAG'].apply(lambda x: x.split('\\')[1]).astype(int)
df_veneziane.drop(columns=['COD_ART_VAR_TIPO_COS_AZ_STDAG','COD_ART_TIPO_COST'], inplace=True)
df_veneziane.rename(columns={'COD_ART':'Codice_articolo','COSTO_ART':'Costo','DATA_RIF':'Data_riferimento'}, inplace=True)
df_veneziane['Categoria'] = 'Veneziane'
df_veneziane['Descrizione'] = ''
df_veneziane = df_veneziane[['Codice_articolo','Codice_variante','Descrizione','Costo','Data_riferimento','Categoria']]
df_veneziane.head()
```

```python
df_vetro = pd.read_excel('data_raw/Aggiornamento_ART_COSTI_per_tipo_articoli_21-08_2023.xlsx', sheet_name='Vetri')
df_vetro['Codice_variante'] = df_vetro['COD_ART_VAR_TIPO_COS_AZ_STDAG'].apply(lambda x: x.split('\\')[1])
df_vetro['Codice_variante'] = df_vetro['Codice_variante'].apply(lambda x: 0 if x == 'ATT' else x).astype(int)
df_vetro.drop(columns=['COD_ART_VAR_TIPO_COS_AZ_STDAG','COD_TIPO_COST','Codice Vetro', 'Costo Vetro', 'Unnamed: 8'], inplace=True)
df_vetro.rename(columns={'COD_ART':'Codice_articolo','COSTO_ART':'Costo','DATA_RIF':'Data_riferimento',"descrizione":"Descrizione"}, inplace=True)
df_vetro['Categoria'] = 'Vetri'
df_vetro = df_vetro[['Codice_articolo','Codice_variante','Descrizione','Costo','Data_riferimento','Categoria']]
df_vetro
```
