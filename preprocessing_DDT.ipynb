```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('data_raw/dbo_DDT_FOR_RIGHE.xlsx', dtype={'COD_CF': object, 'COD_VAR': object})
# remove columns with all 0
df = df.loc[:, (df != 0).any(axis=0)]
# remove columns with all NaN
df = df.dropna(axis=1, how='all')
df["COD_VAR"] = df["COD_VAR"].fillna(0).astype(int)
df
```

```python
df = df.drop(columns=[
    'DOC_RIGA_ID','COD_ART_VAR', 'SCONTO_1', 'SCONTO_2', 'SCONTO_3', 'SCONTO_4',
       'SCONTO_PAG','NOTE','FLAG_CONVERS_EDITABILE', 'FLAG_COSTI_RIGA_MAN','NO_RICALC_PRZ', 'COD_CF_FATT_STDAG', 'FLAG_PERS','FLAG_IMPO_ACQ', 'IVA_INDETRAIBILE_V1', 'NUM_SORT',
       'FLAG_CONVERS_MAN', 'AZIENDA_ID', 'COD_DEP', 'COD_DEP_2','NUM_RIGA','QUANT_RIGA','COD_CAUS_MAG','COSTI_RIGA_V1','NUM_UM',"COD_LIST",	'COD_LIST_VERS',"COD_IVA",
       'PREZZO_NETTO_VU1',"PREZZO_LORDO_VU1","COD_CONTO","IMPORTO_PAG_V1","PREZZO_NETTO_PAG_VU1","CONVERS_UM_BASE","UM"])
#prezzo netto e prezzo lordo sono la stessa cosa
df = df.dropna()
df
```

```python
df = df[["COD_CF","DATA_DOC", "DOC_ID","DES_RIGA","COD_ART","COD_VAR","COD_CAT","UM_BASE","QUANT_UM_BASE","PREZZO_NETTO_TOT_VU1","IMPORTO_V1"]]
df.rename(columns={
    "COD_CF":"COD_CF_Fornitore",
    "DATA_DOC":"Data_documento"
    ,"DOC_ID":"ID_documento",
    "DES_RIGA":"Descrizione",
    "COD_ART":"Codice_articolo",
    "COD_VAR":"Codice_variante",
    "COD_CAT":"Codice_categoria",
    "UM_BASE":"Unita_misura",
    "QUANT_UM_BASE":"Quantita_um",
    "PREZZO_NETTO_TOT_VU1":"Prezzo_netto_vu",
    "IMPORTO_V1":"importo"}, inplace=True)
df
```

```python
# save as csv and xlsx
df.to_csv('data_clean/ddt_righe.csv', index=False)
df.to_excel('data_clean/ddt_righe.xlsx', index=False)
```
